```python
import pandas as pd

train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df
```

```python
test_df
```

```python
train_df.isnull().sum()
```

```python
train_df.dtypes
```

```python
# Fill NA with avg
AVGS = ["Age", "Spa", "VRDeck", "ShoppingMall", "FoodCourt", "RoomService"]
for avg_name in AVGS:
    avg_curr = (train_df[avg_name].mean() + test_df[avg_name].mean()) / 2
    train_df[[avg_name]] = train_df[[avg_name]].fillna(value=avg_curr)
    test_df[[avg_name]] = test_df[[avg_name]].fillna(value=avg_curr)
```

```python
# Fill NA with most popular value
MEANS = ["HomePlanet", "CryoSleep", "Cabin", "Destination", "VIP"]
for column in MEANS:
    max_curr = train_df[column].value_counts().index[0]
    train_df[[column]] = train_df[[column]].fillna(value=max_curr)
    test_df[[column]] = test_df[[column]].fillna(value=max_curr)
```

```python
# Fill Name to -999
train_df.fillna(-999, inplace=True)
test_df.fillna(-999, inplace=True)
```

```python
x = train_df.drop('Transported',axis=1) # train df without survived status
y = train_df.Transported # df with passenger id and survived status
```

```python
x.dtypes
```

```python
# Convert floats to ints, because CatBoost can't use floats
train_df = train_df.astype({"Age": int, "RoomService": int, "FoodCourt": int, "ShoppingMall": int, "Spa": int, "VRDeck": int})
test_df = test_df.astype({"Age": int, "RoomService": int, "FoodCourt": int, "ShoppingMall": int, "Spa": int, "VRDeck": int})

x = x.astype({"Age": int, "RoomService": int, "FoodCourt": int, "ShoppingMall": int, "Spa": int, "VRDeck": int})
```

```python
y
```

```python
import numpy as np
# We will use all indexes after convert
cate_features_index = np.where(x.dtypes != float)[0]
```

```python
cate_features_index
```

```python
from sklearn.model_selection import train_test_split

# Split data to train and valid
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=.85,random_state=42)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(eval_metric='Accuracy',use_best_model=True, random_seed=42, loss_function='Logloss', verbose=True)
```

```python
model.fit(xtrain,ytrain,cat_features=cate_features_index,eval_set=(xtest, ytest))
```

```python
model.predict(xtest).astype('bool')
```

```python
from sklearn.metrics import accuracy_score

print('the test accuracy is :{:.6f}'.format(accuracy_score(ytest,model.predict(xtest).astype('bool'))))
print(accuracy_score(ytrain, model.predict(xtrain).astype('bool')))
```

```python
# Cross val
from catboost import Pool, cv

cv_data = cv(Pool(x,y,cat_features=cate_features_index), model.get_params(),fold_count=10)
```

```python
print('the best cv accuracy is :{}'.format(np.max(cv_data["test-Accuracy-mean"])))
```
